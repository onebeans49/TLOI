```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix

# 데이터 전처리
x_corr = df.drop('OP40S판정', axis=1)
y_corr = df['OP40S판정']

# 상관관계 행렬 계산
corr_matrix = x_corr.corr()

# 피처 중요도 계산
model = DecisionTreeClassifier(random_state=42)
model.fit(x_corr, y_corr)
importances = model.feature_importances_

# 피처 중요도 기반으로 정렬
important_features = np.argsort(importances)[::-1]

# 피처 중요도와 상관관계를 고려한 피처 선택
selected_features = []
for i in important_features:
    if not any([abs(corr_matrix.iloc[i, j]) > 0.8 for j in selected_features]):
        selected_features.append(i)

selected_features_names = x_corr.columns[selected_features]
print("Selected features:", selected_features_names)

# 선택된 피처로 새로운 데이터셋 생성
x_selected = x[selected_features_names]
x_selected_all = x_all[selected_features_names]

# 데이터 분할
x_train_corr, x_test_corr, y_train_corr, y_test_corr = train_test_split(x_selected, y_corr, test_size=0.2, random_state=42)
```

```python
# 모델 학습 및 평가
clf_corr = DecisionTreeClassifier(random_state=42)
clf_corr.fit(x_train_corr, y_train_corr)
y_pred_corr = clf_corr.predict(x_selected_all)

# 성능 평가
accuracy_corr = accuracy_score(y_all, y_pred_corr)
conf_matrix_corr = confusion_matrix(y_all, y_pred_corr)
print(f'Accuracy: {accuracy_corr:.4f}')
print("Confusion Matrix:\n", conf_matrix_corr)
```

# 그리드서치 및 교차검증 (DecisionTree 사용)

```python
from sklearn.model_selection import GridSearchCV, cross_val_score


# 하이퍼파라미터 그리드 정의
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# 그리드 서치 수행 (교차 검증 포함)
grid_search = GridSearchCV(estimator=tree, param_grid=param_grid, cv=5, scoring='f1')
grid_search.fit(x_train_scaled, y_train)

# 최적 하이퍼파라미터 출력
print("Best parameters found: ", grid_search.best_params_)

# 최적 모델로 예측 및 평가
best_tree = grid_search.best_estimator_
y_pred_tree = best_tree.predict(x_test_scaled)

# 성능 평가
accuracy = accuracy_score(y_test, y_pred_tree)
conf_matrix = confusion_matrix(y_test, y_pred_tree)
print(f'Accuracy: {accuracy:.4f}')
print("Confusion Matrix:\n", conf_matrix)
```
